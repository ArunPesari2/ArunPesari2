```

#===============================
# Yield vs Return
#===============================

# Generator-Function : A generator-function is defined like a normal function, but whenever it needs to generate a value, it does so with the yield keyword rather than return. 
# If the body of a def contains yield, the function automatically becomes a generator function.
# generator dont hold entire result in memory
# when the datasets are huge, yield in place of list stores only one value at time rather than all values in the memory
# so yield improves (in memory and time)performance when the data sets are huge.

#======
# yield
#======
def square_numbers(nums):
  for i in nums:
    yield (i*i)

my_nums = square_numbers([1,2,3,4,5])
print(type(my_nums))

print(dir(my_nums))

print(my_nums.__next__())
print(my_nums.__next__())
```
